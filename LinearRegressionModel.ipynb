### __Step 1:__ import necessary libraries


```python
import numpy as np 
from sklearn.linear_model import LinearRegression
```

### __Step 2:__ Now let's define required variables
#### Notice it why and how should we use `reshape` function 

```python
hours_train = np.array([1,2,3,4,5]).reshape(-1,1)#using reshape for making our 1D array to 2D in "1" column
score_train = np.array([50,55,65,70,75])
hours_test = np.array([6,7]).reshape(-1,1)#using reshape for making our 1D array to 2D in "1" column
```

### __Step 3:__
### We initialize the `linear regression model` and fit it to the training data (`hours_train` and `score_train`).

```python
model = LinearRegression()
model.fit(hours_train,score_train)
```

### __Step 4:__
### we use the trained model to predict the scores for the test data `hours_test`, and print the predicted values.


```python
test_Ys = model.predict(hours_test)
print("predicted score is ",test_Ys)
```

### __Step 5:__
### We calculate and print the prediction error for both test values.


```python
first_score_error = test_Ys[0]-80
second_score_error=test_Ys[1]-85
print(f"rediction error for the first tset item is {first_score_error}")
print(f"rediction error for the second tset item is {second_score_error}")
```

### __Optional:__
### you can generate diagrams that can illustrate your process and findings. 

```python
from matplotlib import pyplot as plt


# For Linear Regression (Problem 1)

def plot_linear_regression():
    plt.scatter(hours_train, score_train, color='blue', label='Training Data')
    plt.plot(hours_test, test_Ys, color='red', label='Predicted Score')
    plt.xlabel('Hours Studied')
    plt.ylabel('Score')
    plt.legend()
    plt.title('Linear Regression: Hours vs Score')
    plt.savefig('images/linear_regression.png')
    plt.show()
plot_linear_regression()
```
